```python
import os
import sys

import numpy as np
from torchvision import transforms

sys.path.append("../")
from src import create_hf_train_dataset, FashionNetDataset, create_attribute_dict, show_tensor_images
```

```python
data_dir = "../data"
```

```python
attribute_dict = create_attribute_dict(os.path.join(data_dir, "split/list_attr_cloth.txt"))
```

```python
dash_line = "-".join(" " for _ in range(100))
```

```python
def visualize_data(data):
    image, label_list = data
    attributes = [attribute_dict[i][label] for i, label in enumerate(label_list)]
    
    print(dash_line)
    print(f"IMAGE:\n")
    show_tensor_images(image.unsqueeze(0), num=1)
    print(dash_line)
    print(f"LABEL:\n{label_list}")
    print(dash_line)
    print(f"ATTRIBUTES:\n{attributes}")
    print(dash_line)
```

```python
hf_dataset_dict = create_hf_train_dataset(data_dir)
```

```python
transforms = transforms.Compose([transforms.ToTensor()])
train_dataset = FashionNetDataset(hf_dataset_dict["train"], transforms)
```

```python
example_indices = np.random.randint(0, len(train_dataset), 5)

for i, index in enumerate(example_indices):
    print(dash_line)
    print("Example", i+1)
    visualize_data(train_dataset[int(index)])
```
