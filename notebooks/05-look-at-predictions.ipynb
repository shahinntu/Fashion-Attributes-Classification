```python
import os
import sys

import torch

sys.path.append("../")
from src import create_hf_test_dataset, Predictor, save_list_to_txt
```

```python
data_dir = "../data"
model_log_dir = "../model_logs"
restore_version = "train_240319135530"
results_dir = "../data/results"
```

```python
test_dataset_dict = create_hf_test_dataset(data_dir)
```

```python
predictor = Predictor.from_pretrained(model_log_dir, restore_version, device)
```

```python
predictions = predictor.predict(test_dataset_dict["test"])
```
